```python
import gymnasium as gym
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import defaultdict
import math
```

```python
def n_step_TD_prediction(env, policy, n, step_size=0.5, discount_factor=0.99, episodes=10000):
    nS = env.observation_space.n
    nA = env.action_space.n
    V = np.zeros(nS)
    for i in range(episodes):
        state, _ = env.reset()
        T = math.inf
        t = 0

        rewards = [0.0] * (n+1)
        states = [0] * (n+1)
        states[0] = state
        while(True):
            if t < T:
                action = np.random.choice(nA, p=policy[state])
                s_next, reward, term, trun, _ = env.step(action)
                states[(t+1) % (n+1)] = s_next #store S(t+1)
                rewards[(t+1) % (n+1)] = reward #store R(t+1)
                state = s_next
                if term or trun:
                    T = t+1
            tau = t - n + 1 #the time whose state's estimate is being updated 
            if tau >= 0: #once we have enough rewards to make updates (after n time-steps)
                upper = min(tau + n, T) #this is for if the episode terminates in the middle of the sliding window
                G = 0.0
                for k in range(tau + 1, upper + 1): #include the last one
                    G += (discount_factor ** (k - tau - 1)) * rewards[k % (n + 1)]
                if tau+n < T: #if episode terminates in sliding window, we cant bootstrap since we cant bootstrap with a terminal state
                    G += (discount_factor ** n) * V[states[(tau + n) % (n + 1)]]
                V[states[tau % (n + 1)]] += step_size * (G - V[states[tau % (n + 1)]])
            t+=1
            if tau == T - 1:
                break
    return V
```
